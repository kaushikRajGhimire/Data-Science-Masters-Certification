#Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

Class: A class is a blueprint or template for creating objects. It defines the properties (attributes) and behaviors (methods) that the objects created from the class will have.

Object: An object is an instance of a class. It represents a real-world entity or concept that the class defines.


class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def display_info(self):
        print(f"Car Brand: {self.brand}, Model: {self.model}")

# Creating objects (instances) of the Car class
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

car1.display_info()  # Output: Car Brand: Toyota, Model: Corolla
car2.display_info()  # Output: Car Brand: Honda, Model: Civic
In this example:

Car is the class (blueprint).
car1 and car2 are objects (instances) of the Car class.

#Q2. Name the four pillars of OOPs.
The four pillars of Object-Oriented Programming (OOP) are:

Encapsulation: Bundling data (attributes) and methods that operate on the data into a single unit, i.e., a class. It also provides access control through access modifiers (e.g., private, protected, public).

Abstraction: Hiding the complex implementation details and exposing only the necessary functionalities to the users.

Inheritance: The mechanism by which one class acquires the properties and behaviors of another class.

Polymorphism: The ability of a single function, method, or operator to behave differently based on the input or object. It can be achieved via method overriding and overloading.

#Q3. Explain why the __init__() function is used. Give a suitable example.
__init__() is a special method in Python known as the constructor. It is automatically called when an object is created from a class, and it allows the class to initialize the object's attributes.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating an object of Person class
person1 = Person("Alice", 25)
person1.display_info()  # Output: Name: Alice, Age: 25
In this example:

__init__() initializes the name and age attributes for each object created from the Person class.
When person1 is created, __init__() automatically sets the values for name and age.

#Q5. What is inheritance? Give an example for each type of inheritance in Python.
Inheritance is a mechanism in which a new class (derived/child class) acquires the properties and behaviors (methods and attributes) of an existing class (base/parent class). This allows code reuse and helps in hierarchical classification.

Types of Inheritance:
#Single Inheritance: A derived class inherits from a single base class.


class Animal:
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):  # Single Inheritance
    def sound(self):
        print("Bark")

dog = Dog()
dog.sound()  # Output: Bark

Multiple Inheritance: A derived class inherits from more than one base class.

    def fly(self):
        print("Flying")

Multilevel Inheritance: A derived class inherits from another derived class.

class Vehicle:
    def move(self):
        print("Vehicle is moving")

class Car(Vehicle):  # Multilevel Inheritance

Hierarchical Inheritance: Multiple derived classes inherit from the same base class.

class Shape:
    def area(self):
        pass

class Circle(Shape):  # Hierarchical Inheritance
    def area(self, radius):
        return 3.14 * radius * radius

class Square(Shape):
    def area(self, side):
        return side * side

circle = Circle()
print(circle.area(5))  # Output: 78.5

square = Square()
print(square.area(4))  # Output: 16


Hybrid Inheritance: A combination of more than one type of inheritance.

class A:
    def methodA(self):
        print("Method A")

class B(A):  # Single Inheritance
    def methodB(self):
        print("Method B")

class C:
    def methodC(self):
        print("Method C")

class D(B, C):  # Multiple Inheritance
    def methodD(self):
        print("Method D")

obj = D()
obj.methodA()  # Output: Method A
obj.methodB()  # Output: Method B
obj.methodC()  # Output: Method C
obj.methodD()  # Output: Method D
